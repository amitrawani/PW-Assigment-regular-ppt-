Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating 
visually appealing and informative statistical graphics. Here are five commonly used plots in Seaborn and their uses:

1. Scatter Plot:

Uses: Scatter plots are used to visualize the relationship between two continuous variables. They can show patterns, correlations, 
or clusters in the data.

2. Line Plot:

Uses: Line plots are useful for displaying trends or changes over time. They are commonly used for time series data or to show the 
relationship between variables with a continuous progression.

3. Bar Plot:

Uses: Bar plots are effective for comparing categories or groups. They represent the distribution or frequency of categorical data 
and can reveal patterns, differences, or rankings between the groups.

4. Histogram:

Uses: Histograms display the distribution of a single variable by dividing the data into bins and showing the frequency or density 
of values within each bin. They provide insights into the shape, spread, and central tendency of the data.

5. Heatmap:

Uses: Heatmaps are used to visualize the matrix-like data where values are represented by colors. They are useful for
displaying correlations, patterns, or clustering in large datasets and are commonly used in exploratory data analysis.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using 

x ="timepoint" and 

y = "signal" for different events and regions.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Plot line plot for different events and regions
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Set title and labels
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot box plots for 'pclass' and 'age'
plt.figure(figsize=(10, 4))  # Set the figure size

plt.subplot(1, 2, 1)  # Create the first subplot
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title("Box Plot of Age by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Age")

plt.subplot(1, 2, 2)  # Create the second subplot
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title("Box Plot of Fare by Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot histogram with hue for 'cut'
sns.histplot(data=diamonds_data, x='price', hue='cut')

# Set title and labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot pair plot with hue for 'species'
sns.pairplot(data=iris_data, hue='species')

# Display the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix
flights_matrix = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap
sns.heatmap(flights_matrix, cmap="YlGnBu")

# Set title and labels
plt.title("Flights Passengers Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
