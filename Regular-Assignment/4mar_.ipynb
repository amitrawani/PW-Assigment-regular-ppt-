# Plotly Assignment

```python
import plotly.express as px
import seaborn as sns
```

Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
titanic = sns.load_dataset("titanic")
px.scatter(data_frame=titanic, x="age",y="fare",title="TITANIC AGE VS FARE")


```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
tips = sns.load_dataset("tips")
px.box(data_frame=tips,x="total_bill",title="BOXPLOT OF TOTAL BILL FROM TIPS")
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. 
Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
px.histogram(data_frame=tips,x="sex",y="total_bill",pattern_shape="smoker",color="day",title = "TIPS DATASET")
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.


```python
iris = sns.load_dataset("iris")
px.scatter_matrix(data_frame=iris,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color="species")
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

```python
px.histogram(data_frame=tips,x="total_bill",y="tip",color="sex",marginal="box",title="TIPS DATA")
```
