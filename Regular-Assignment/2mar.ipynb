Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python.
It provides a wide range of functions and classes for generating various types of plots, graphs, and charts.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: This plot is used to display data points connected by straight lines. It is commonly used to visualize trends over time or
to show the relationship between two variables.

2. Scatter Plot: Scatter plots are used to display the relationship between two continuous variables. Each data point is represented by 
a marker on the plot, and the position of the marker represents the values of the two variables.

3. Bar Plot: Bar plots are used to compare different categories or groups. They display rectangular bars with lengths proportional to the
values they represent. Bar plots are useful for visualizing categorical data and comparing quantities across different groups.

4. Histogram: Histograms are used to represent the distribution of a continuous variable. It divides the data range into bins or intervals
and displays the number of data points falling into each bin. Histograms are commonly used to analyze the frequency or density of data.

5. Pie Chart: Pie charts represent data as slices of a circular pie, where each slice corresponds to a specific category or proportion of 
a whole. Pie charts are useful for showing the relative sizes or percentages of different categories within a dataset


Pyplot module of Matplotlib. Matplotlib offers many more types of plots, including area plots, box plots, violin plots, and more, allowing y to create a wide range of visualizations to suit our needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot used to display the relationship between two continuous variables. Each data point is 
represented by a marker on the plot, and the position of the marker corresponds to the values of the two variables being compared.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)

# Adding title, xlabel, and ylabel
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")

# Display the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
    
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans: The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display 
multiple plots in a single figure, making it easier to compare and analyze different data or visualizations.

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plotting line 1
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title("Line 1")

# Plotting line 2
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title("Line 2")

# Plotting line 3
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title("Line 3")

# Plotting line 4
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title("Line 4")

# Adjusting the spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans: A bar plot is a type of plot that uses rectangular bars to represent categorical data. It is commonly used to compare and 
visualize the quantities or values associated with different categories or groups.

A bar plot is used for the following purposes:

1. Comparison: It allows you to compare the values or quantities of different categories or groups easily. The length or height of 
each bar represents the value being compared.

2. Ranking: Bar plots can be used to rank categories based on their values. The tallest bar indicates the highest value, while the
shortest bar indicates the lowest value.

3.Distribution: Bar plots can also be used to represent the distribution of categorical data by grouping categories and displaying 
the frequencies or proportions of each group.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit ($)")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit ($)")
plt.ylabel("Company")
plt.show()
```
